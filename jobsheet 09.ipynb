# Praktikum 01 - Menggunakan Modul math

```python
import math

angka = 16
hasil1 = math.sqrt(angka)
hasil2 = math.pow(2, 5)
print("Akar kuadrat dari", angka, "=", hasil1)
print("2 pangkat 5 =", hasil2)
```

# Penugasan 01 - Eksperimen Modul math

```python
import math

def luas_lingkaran(r):
    luas = math.pi * (r ** 2)
    return luas

r1 = 7
r2 = 10

print(f"Luas lingkaran dengan r = {r1} yaitu: {luas_lingkaran(r1):.2f}")
print(f"Luas lingkaran dengan r = {r2} yaitu: {luas_lingkaran(r2):.2f}")
```

# Penugasan 02 - Eksperimen dengan Modul random

```python
import math
import random

print("Angka acak dari 1-100 =")

for i in range(5):
    angka = random.randint(1, 100)
    print(angka)
```
