# Praktikum 01 - Menggunakan Modul math

```python
import math

angka = 16
hasil1 = math.sqrt(angka)
hasil2 = math.pow(2, 5)
print("Akar kuadrat dari", angka, "=", hasil1)
print("2 pangkat 5 =", hasil2)
```

# Praktikum 02 - Menggunakan from....import...

```python
from math import sqrt

angka = 25
hasil = sqrt(angka)
print("Akar kuadrat dari", angka, "=", hasil)
```

# Praktikum 03 - Memberi Alias Modul

# Praktikum 04 - Eksplorasi math, random, dan sys

```python
import math
import random
import sys

hasil1 = math.sqrt(81)
hasil2 = random.randint(1, 10)
versi = sys.version

print("Akar kuadrat 81 =", hasil1)
print("Angka acak 1-10 =", hasil2)
print("Versi Python =", versi)
```
