# Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

## Praktikum 02 - Selection Sort

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

# Praktikum 03 - Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 04 - Studi Kasus: Mengurutkan Nilai Mahasiswa


```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1] [1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

# Praktikum 05 - Perbandingan Waktu Ekskusi Sorting

```python
# --- Membuat file bubble_sort.py ---
with open("bubble_sort.py", "w") as f:
    f.write("""
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
""")

# --- Membuat file selection_sort.py ---
with open("selection_sort.py", "w") as f:
    f.write("""
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
""")

# --- Membuat file insertion_sort.py ---
with open("insertion_sort.py", "w") as f:
    f.write("""
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
""")

print("Semua modul berhasil dibuat.")
```

```python
import random, time
from bubble_sort import bubble_sort
from selection_sort import selection_sort
from insertion_sort import insertion_sort

data = [random.randint(1, 1000) for _ in range(500)]

for func in [bubble_sort, selection_sort, insertion_sort]:
    salinan = data.copy()
    start = time.time()
    func(salinan)
    end = time.time()
    print(f"{func.__name__} membutuhkan waktu {end - start:.5f} detik")
```

# Praktikum 06 - Sorting Data Barang(Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1] [key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1] [key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]

print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```
