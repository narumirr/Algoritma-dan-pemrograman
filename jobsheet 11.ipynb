# Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}"
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

# Praktikum 02 - Binary Search pada Data Angka

```python
def binary_search(data, target):
    left, right =0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid -1
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

# Praktikum 03 - Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
    for i in range(len(data)):
        if data[i].lower() == target.lower():
            return f"Nama {target} ditemukan di indeks {i}"
    return f"Nama {target} tidak ditemukan"
nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

# Praktikum 04 - Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right)  // 2
        if data[mid] == target:
            return f"Nilai {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

# Praktikum 05 - Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

## Praktikum 06 - Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j>= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid]  == target:
            return f"{target} ditemukan di indeks {mid}"
        elif data[mid]  < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7]
print(binary_search(urut, target))
```

# Variasi Sequential Search

```python
def sequential_search_count(data, target):
    count = 0  # untuk menghitung jumlah perbandingan
    for i in range(len(data)):
        count += 1   # setiap pengecekan dihitung
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i} setelah {count} kali perbandingan"
    return f"Data {target} tidak ditemukan setelah {count} kali perbandingan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search_count(angka, 30))
print(sequential_search_count(angka, 100))
```

# Variasi Binary Search

```python
def binary_search_steps(data, target):
    left, right = 0, len(data) - 1
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        if data[mid] == target:
            return mid, steps
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1, steps

angka = [10, 20, 30, 40, 50]
target = 40
index, steps = binary_search_steps(angka, target)
print("Data:", angka)
print("Target", target)
print("Index ditemukan:", index)
print("Jumlah langkah:", steps)
```

# Studi Kasus Barang Toko

```python
def binary_search_barang(data, target_harga):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid][1] == target_harga:
            return mid
        elif data[mid][1] < target_harga:
            left = mid + 1
        else:
            right = mid - 1
    return -1
# Input daftar barang
n = int(input("Jumlah barang: "))
barang = []
for _ in range(n):
    nama = input("Nama barang: ")
    harga = int(input("Harga: "))
    barang.append((nama, harga))
# Sort berdasarkan harga ASCENDING
barang.sort(key=lambda x: x[1])
print("\nData setelah diurutkan berdasarkan harga:")
print(barang)
target = int(input("\nMasukkan harga yang ingin dicari: "))
index = binary_search_barang(barang, target)
if index != -1:
    print(f"Barang dengan harga {target} ditemukan: {barang[index]}")
else:
    print("Barang tidak ditemukan.")
```

# Analisis Kompleksitas

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


sizes = [100, 1000, 10000]
print("Ukuran | Seq Time (s) | Bin Time (s)")
print("--------------------------------------")
for size in sizes:
    data = list(range(size))
    target = size - 1
    # sequential time
    start = time.time()
    sequential_search(data, target)
    seq_time = time.time() - start
    # binary time
    start = time.time()
    binary_search(data, target)
    bin_time = time.time() - start
    print(f"{size:<6} | {seq_time:<12.8f} | {bin_time:<12.8f}")
```
