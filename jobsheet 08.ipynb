# Praktikum 04 - Fungsi dengan *args dan ** kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 - Fungsi Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan....")
        fungsi()
        print("Setelah fungsi dijalankan.....")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

# Praktikum 08 - Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

```python
def hitung_rata(data):
    return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("Rata-rata nilai:", rata)
```

# Penugasan 04 - Fungsi dengan args

```python
def jumlahkan_semua(*angka):
    return sum(angka)

print("Jumlah 3 bilangan = ", jumlahkan_semua(10, 20, 30))
print("Jumlah 5 bilangan = ", jumlahkan_semua(1, 2, 3, 4, 5))
print("Jumlah 7 bilangan = ", jumlahkan_semua(7, 6, 5, 4, 3, 2, 1))
```

# Penugasan 05 - Fungsi dengan kwargs

```python
def info_barang(**data):
    for key, value in data.items():
        print(f"{key}: {value}")

info_barang(Nama="Shotgun", Harga=5000000, Stok=25)
print()
info_barang(Nama="Ak-47", Harga=7500000, Stok=37)
```

# Penugasan 08 - Fungsi Decorator

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, Mahasiswa Polines!")

halo()
```

# Penugasan 09 - Studi Kasus Nyata - Program Nilai

```python
# a. fungsi menerima nilai mahasiswa(list)
def input_nilai():
    nilai_tugas = float(input("Masukkan nilai tugas: "))
    nilai_ujian = float(input("Masukkan nilai ujian: "))
    return [nilai_tugas, nilai_ujian]

# b. menghitung rata-rata nilai
def hitung_rata(daftar_nilai):
    return sum(daftar_nilai) / len(daftar_nilai)

# c. menentukan apakah mahasiswa lulus (>=60) atau tidak
def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

# d. labda menghitung nilai akhir berdasarkan bobot
nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

print("== Program Penilaian Mahasiswa ===")
tugas, ujian = input_nilai()

akhir = nilai_akhir(tugas, ujian)
rata = hitung_rata([tugas, ujian])
status = status_kelulusan(akhir)

print("\n--- Hasil ---")
print(f"Nilai Tugas : {tugas}")
print(f"Nilai Ujian : {ujian}")
print(f"Nilai Akhir : {akhir:.2f}")
print(f"Rata-rata   : {rata:.2f}")
print(f"Status      : {status}")
```
